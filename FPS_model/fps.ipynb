```python
app = Flask(__name__)
#모델 불러오기



    

    
@app.route('/')
def model():
    K.clear_session()
    
    value = request.args.get('fname')
    resources = urllib.request.urlopen('http://localhost:8081/FPS/img/'+value)
    image = Image.open(resources)
    
    image.save('C:/Users/sm-02/.conda/envs/YOLOdark/darkflow/Mask_RCNN/imgfile/'+value)
    
    image_path="C:/Users/sm-02/.conda/envs/YOLOdark/darkflow/Mask_RCNN/imgfile/"+value
    image = cv2.imread(image_path)
    plt.imshow(image)
    image1 = image
    
    t = time.time()

    # 모델 적용
    config=balloon.BalloonConfig()

    models = modellib.MaskRCNN(mode="inference", config=config, model_dir='logs/mask_rcnn_hand_0031.h5')
    models.load_weights("logs/mask_rcnn_hand_0031.h5", by_name=True)

        # 2.
    protoFile = "C:/Users/sm-02/.conda/envs/YOLOdark/darkflow/Mask_RCNN/hand/pose_deploy.prototxt"
    weightsFile = "C:/Users/sm-02/.conda/envs/YOLOdark/darkflow/Mask_RCNN/hand/pose_iter_102000.caffemodel"
    nPoints = 22
    POSE_PAIRS = [ [0,1],[1,2],[2,3],[3,4],[0,5],[5,6],[6,7],[7,8],[0,9],[9,10],[10,11],[11,12],[0,13],[13,14],[14,15],[15,16],[0,17],[17,18],[18,19],[19,20] ]
    net = cv2.dnn.readNetFromCaffe(protoFile,weightsFile)
    
    
    result = balloon.detect_and_color_splash(model=models,image_path=image_path,video_path=None)
    print(2)
    print("time taken by network : {:.3f}".format(time.time() - t))
    
    time1 = time.time() - t
    

    for rois in (result['rois']):
        x_0 = rois[0]
        y_0 = rois[1]
        x_1 = rois[2]
        y_1 = rois[3]
        x0=(int)(x_0*0.9)
        y0=(int)(y_0*0.9)

        x1=(int)(x_1*1.1)
    #     if x1>image.shape[1] :
    #         x1=image.shape[1]
    #         print(x1)
        y1=(int)(y_1*1.1)

    #     if y1>image.shape[0] :
    #         y1=image.shape[0]
    #         print(y1)
        frame = image[x0:x1, y0:y1]

        frameOrg = np.copy(frame)
        frameCopy = np.copy(frame)
        frameWidth = frame.shape[1]
        frameHeight = frame.shape[0]
        aspect_ratio = frameWidth/frameHeight
        threshold = 0.1

        # input image dimensions for the network
        inHeight = 368
        # inWidth = x1-x0
        inWidth = int(((aspect_ratio*inHeight)*8)//8)
        inpBlob = cv2.dnn.blobFromImage(frame, 1.0 / 255, (inWidth, inHeight),
                                (0, 0, 0), swapRB=False, crop=False)

        net.setInput(inpBlob)

        output = net.forward()


        points = []
        

        for i in range(nPoints):
            # confidence map of corresponding body's part.
            probMap = output[0, i, :, :]
            probMap = cv2.resize(probMap, (frameWidth, frameHeight))

            # Find global maxima of the probMap.|
            minVal, prob, minLoc, point = cv2.minMaxLoc(probMap)

            if prob > threshold :
                cv2.circle(frameCopy, (int(point[0]), int(point[1])), 3, (0, 255, 255), thickness=-1, lineType=cv2.FILLED)
                cv2.putText(frameCopy, "{}".format(i), (int(point[0]), int(point[1])), cv2.FONT_HERSHEY_SIMPLEX, .5, (0, 0, 255), 2, lineType=cv2.LINE_AA)

              # Add the point to the list if the probability is greater than the threshold
                points.append((int(point[0]), int(point[1])))
            else :
                points.append(None)


#         plt.figure(figsize=[10,10])
#         plt.imshow(cv2.cvtColor(frameCopy, cv2.COLOR_BGR2RGB))

        cv2.imwrite(str(rois[0])+"12321_catch.jpg",frameCopy)
        


    #     plt.imshow(frameCopy)

        delpoint1 = points[3]
        delpoint2 = points[4]
        delpoint3 = points[7]
        delpoint4 = points[8]
        delpoint5 = points[11]
        delpoint6 = points[12]
        delpoint7 = points[15]
        delpoint8 = points[16]
        delpoint9 = points[19]
        delpoint10 = points[20]

        del_list = []
        del_list.append(delpoint1)
        del_list.append(delpoint2)
        del_list.append(delpoint3)
        del_list.append(delpoint4)
        del_list.append(delpoint5)
        del_list.append(delpoint6)
        del_list.append(delpoint7)
        del_list.append(delpoint8)
        del_list.append(delpoint9)
        del_list.append(delpoint10)
        img = frameOrg




        for i in range(0, 9, 2):
            try:            
                x2 = (int)((del_list[i][0]+del_list[i+1][0])/2)
                y2 = (int)((del_list[i][1]+del_list[i+1][1])/2)
                d1 = (int)(abs(del_list[i][0]-del_list[i+1][0])/2)
                d2 = (int)(abs(del_list[i][1]-del_list[i+1][1])/2)

                d3=(int)((d1+d2)*0.75)
                r=5

                print(x2,y2,d3)

                for i in range(r):
                    hand = img[y2-d3:y2+d3, x2-d3:x2+d3]
                #        blur = cv2.GaussianBlur(hand, (5, 5), (int)((d3-i)/r))
                    blur = cv2.GaussianBlur(hand, (9, 9), 101)
                      # src: 원본 이미지
                      # ksize: 커널 크기
                      # sigmaX: 표준편차
                    img[y2-d3:y2+d3, x2-d3:x2+d3] = blur 
            except:
                print("cantcatch")
        plt.figure(figsize=[10,10])
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

        cv2.imwrite(str(rois[0])+"_converted.jpg",img)

        image[x0:x1, y0:y1] = img
    
    img_final = image
    plt.imshow(img_final)
    
    cv2.imwrite("inho.jpg", image)
    print("time taken by network : {:.3f}".format(time.time() - t))
    time2 = time.time() - t
    
    height = image1.shape[0]
    width = image1.shape[1]
    td = time1 + time2
    print(td)
    
    

    
    

    img_final = Image.fromarray(img_final)
    img_final.save('C:/Users/sm-02/Desktop/3_PROJECT_FPS_Web/3차프로젝트_FPS_Web/FPS/WebContent/img2/'+value)
    time.sleep(3)
    return redirect('http://localhost:8081/FPS/main.jsp?h='+str(height)+'&w='+str(width)+'&t='+str(td))
    
if __name__ == '__main__':
    app.run(host='localhost', port='9000')
```
